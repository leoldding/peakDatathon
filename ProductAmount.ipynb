```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use('fivethirtyeight')
```

```python
orders = pd.read_csv('data/olist_orders_dataset.csv')
order_items = pd.read_csv('data/olist_order_items_dataset.csv')
products = pd.read_csv('data/olist_products_dataset.csv')
```

## Product Amount - Analysis

```python
orderItemsMerged = pd.merge(orders,order_items, on='order_id')
overall = pd.merge(orderItemsMerged, products, on='product_id')
```

```python
# looking for null values
overall.isnull().sum()
```

```python
# checking specific orders 
overall.loc[overall['order_approved_at'].isnull()]
```
